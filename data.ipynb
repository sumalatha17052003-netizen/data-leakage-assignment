```python
df.dropna(subset="selling_price",inplace=True)
print(df["selling_price"].isnull().sum())
```

```python
df["mileage"]=df["mileage"].str.strip().str.title()
df["mileage"].unique()
```

```python
df["mileage"]=df["mileage"].astype(str).str.replace("Kmpl","")
df["mileage"]
```

```python
df["mileage"]=df["mileage"].astype(str).str.replace("Km/L","")
df["mileage"]
```

```python
df["mileage"]=pd.to_numeric(df["mileage"],errors="coerce")
df["mileage"].isnull().sum()
```

```python
df["mileage"]=df["mileage"].fillna(df["mileage"].mean())
df["mileage"].isnull().sum()
```

```python
df["mileage"]=pd.to_numeric(df["mileage"],errors="coerce")
df["mileage"]
```

```python
df["engine"].isnull().sum()
```

```python
df["engine"]=df["engine"].fillna(df["engine"].mean())
df["engine"].isnull().sum()
```

```python
df.info()
```

```python
df["max_power"].isnull().sum()
df["max_power"]=df["max_power"].fillna(df["max_power"].median())
df["max_power"].isnull().sum()
df
```

```python
df.drop(columns=["Unnamed: 0"], inplace=True)
df

```

```python
typo={"Manual":0,"Automatic":1}
df["transmission_type"]=df["transmission_type"].replace(typo)
df["transmission_type"]
```

```python
df["fuel_type"].unique()
```

```python
map={ 'Diesel':'DIESEL',
     ' Diesel':'DIESEL',
      'DIESEL':'DIESEL',
'diesel':'DIESEL',
'diesel ':'DIESEL',
'Petrol':'PETROL',
'Petrol ':'PETROL',
' Petrol':'PETROL',
'petrol ':'PETROL',
'PETROL':'PETROL',
'petrol':'PETROL',
     'CNG':'CNG',
     'LPG':'LPG',
     'Electric':'ELECTRIC'}
df["fuel_type"]=df["fuel_type"].replace(map)
df["fuel_type"].unique()
```

```python
fuel_dummies = pd.get_dummies(df["fuel_type"], drop_first=True)
df = pd.concat([df, fuel_dummies], axis=1)
df["fuel_type"]
```

```python
X = df.drop(columns=["selling_price"])
y = df["selling_price"]
X = X.select_dtypes(include=["number"])
X
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test,y_train,y_test = train_test_split(
X,y,test_size=0.2,random_state=43)
```

```python
import numpy as np
from sklearn.metrics import mean_absolute_error
y_pred = np.full_like(y_test, y_train.mean(), dtype=np.float64)
baseline_mae = mean_absolute_error(y_test, y_pred)
print("Baseline MAE:", baseline_mae)
```
