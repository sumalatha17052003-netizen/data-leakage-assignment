```python
import pandas as pd
df=pd.read_csv("used_cars_messy.csv")
print(df.head(10))
print(df.info())
print(df.describe())
print(df.shape)
df.drop_duplicates(inplace=True)
df
```

```python
df['selling_price'].isnull().sum()
df['selling_price']=df['selling_price'].fillna(df['selling_price'].mean(), inplace=True)
df['selling_price'].isnull().sum()
import pandas as pd
from sklearn.metrics import mean_absolute_error
mean_price = df["selling_price"].mean()
baseline_preds = [mean_price] * len(df)
mae = mean_absolute_error(df["selling_price"], baseline_preds)
print("Baseline MAE:", mae)


```
